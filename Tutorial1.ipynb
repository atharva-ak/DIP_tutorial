```python
#Original image
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('lenna.png')  # load image
img1 = cv2 . cvtColor(img,cv2 . COLOR_BGR2RGB)
plt.imshow(img1)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# Grayscale image

img = cv2.imread('lenna.png',0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# R,G,B Channels
img = cv2.imread('lenna.png')
img1 = cv2 . cvtColor(img,cv2 . COLOR_BGR2RGB)
# g = img1.copy()
# set blue and red channels to 0
r= img1[:, :, 0]
g= img1[:, :, 1]
b= img1[:, :, 2]



plt.figure(1)
plt.subplot(131)
plt.imshow(r, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])

plt.subplot(132)
plt.imshow(g, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])

plt.subplot(133)
plt.imshow(b, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Addition of two images
import cv2 
import numpy as np
  
# Read Image1 
image1 = cv2.imread('lenna.png',0) 
  
# Read image2 
image2 = cv2.imread('got.jpg',0) 
image2 = cv2.resize(image2, (220, 220))
  
# Add the images 
img = cv2.add(image1, image2) 

# Subtract images
# img = np.int32(img)-np.int32(image2) 

# Show the image 
plt.imshow(img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
import numpy as np
import cv2

# Create a black image
img = cv2.imread('lenna.png')
img = cv2.rectangle(img,(55,55),(150,150),(255,255,255), -1)
plt.imshow(img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show() 
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.imread('lenna.png')
img = cv2.putText(img,'OpenCV',(55,100), font, 1,(255,255,0),2,cv2.LINE_AA)
plt.imshow(img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show() 
```
